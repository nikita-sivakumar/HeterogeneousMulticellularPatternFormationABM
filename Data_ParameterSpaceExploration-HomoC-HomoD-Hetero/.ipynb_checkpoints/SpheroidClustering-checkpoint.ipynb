```python
import pandas as pd
import numpy as np
from os import listdir
from matplotlib import image
import PIL
from PIL import Image
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X = []
X = pd.read_excel('features_spheroids.xlsx')
X.shape
```

# (25, 25) clustering:

```python
X_25_25 = X[(X["homo_c"] == 25) & (X["homo_d"] == 25) & (X["hetero"] == 0)]
```

```python
inertia=[]
for i in range(2,10):
    kmeans = KMeans(n_clusters=i, random_state=0,n_init=25).fit(
        X_25_25[["num_green_regions","num_red_regions"]])
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(np.array(range(2,10)), inertia)
```

```python
kmeans_25_25_4 = kmeans = KMeans(n_clusters=4, random_state=0,n_init=25).fit(
        X_25_25[["num_green_regions","num_red_regions"]])
```

```python
import random
import matplotlib.pyplot as plt
X_25_25["label"] = kmeans_25_25_4.labels_
fig=plt.figure(figsize=(10,10))
c = 0
for i in range(4):
    group = X_25_25[X_25_25["label"] == i]
    group_sample=np.random.choice(np.array(group["file-path"]),10,replace=False)    
    for j in group_sample:
        c = c+1
        img = Image.open(j)
        rgb = img.convert('RGB')
        fig.add_subplot(8, 10, c)
        plt.imshow(rgb)
```

```python
print(sum(X_25_25["label"]==0))
print(sum(X_25_25["label"]==1))
print(sum(X_25_25["label"]==2))

kmeans_25_25_4.cluster_centers_
```

```python
plt.scatter(X_25_25["num_green_regions"],X_25_25["num_red_regions"],alpha=0.1)
plt.scatter([i[0] for i in kmeans_25_25_4.cluster_centers_],[i[1] for i in kmeans_25_25_4.cluster_centers_])
```

```python
plt.scatter(X_25_25[X_25_25["label"]==0]["num_green_regions"],X_25_25[X_25_25["label"]==0]["num_red_regions"],alpha = 0.2,c="orange")
plt.scatter(X_25_25[X_25_25["label"]==1]["num_green_regions"],X_25_25[X_25_25["label"]==1]["num_red_regions"],alpha = 0.2,c="red")
plt.scatter(X_25_25[X_25_25["label"]==2]["num_green_regions"],X_25_25[X_25_25["label"]==2]["num_red_regions"],alpha = 0.2,c="blue")
plt.scatter(X_25_25[X_25_25["label"]==3]["num_green_regions"],X_25_25[X_25_25["label"]==3]["num_red_regions"],alpha = 0.2,c="green")
plt.xlabel("Number of Green Regions")
plt.ylabel("Number of Red Regions")
```

```python
h = 0.02

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = X_25_25["num_green_regions"].min() - 1, X_25_25["num_green_regions"].max() + 1
y_min, y_max = X_25_25["num_red_regions"].min() - 1, X_25_25["num_red_regions"].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans_25_25_4.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = np.array(Z).reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot( X_25_25["num_green_regions"], X_25_25["num_red_regions"], 'k.', markersize=5,alpha=0.1)
# Plot the centroids as a white X
centroids = kmeans_25_25_4.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on X=25, Y=25')
# plt.xlim(x_min, x_max)
# plt.ylim(y_min, y_max)
# plt.xticks(())
# plt.yticks(())
plt.show()
```

# Clustering HomoC/HomoD Spheroids

```python
import pandas as pd
import numpy as np
from os import listdir
from matplotlib import image
import PIL
from PIL import Image
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X = []
X = pd.read_excel('features_spheroids.xlsx')
X.shape
```

```python
X_subset = X[(X["hetero"] == 0)]
```

```python
X_subset.columns
```

```python
inertia=[]
for i in range(2,20):
    kmeans = KMeans(n_clusters=i, random_state=0,n_init=25).fit(
        X_subset[["num_green_regions","num_red_regions", 
                  "green_region_aspect_ratio","red_region_aspect_ratio","contiguous_area",
                  "green_avg_regionArea_fract","green_avg_cent_dist",
                  "red_avg_regionArea_fract","red_avg_cent_dist"]])
    inertia.append(kmeans.inertia_)
```
